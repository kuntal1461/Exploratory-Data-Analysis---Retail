# EDA

```python
sns.pairplot(data=df)
```

```python
df["Ship Mode"].unique()
```

```python
#Quantity Vs Profits

sns.barplot(df["Quantity"],df["Profit"],ci=None)
plt.title("Quantity Vs Profits",fontsize=30)
plt.show()
```

```python
# Ship Mode Vs Profit 
sns.barplot(df["Ship Mode"],df["Profit"],ci=None)
plt.title("Ship Mode Vs Profit",fontsize=30)
plt.show()
```

```python
len(df["State"].unique())
```

```python
# Profits Accoding The Sates

order=df.groupby(["State"])["Profit"].mean().sort_values().index
plt.figure(figsize=(20,30),dpi=400)
sns.barplot(df["Profit"],df["State"],ci=None,order=order)
plt.title("Profits Accoding The Sates",fontsize=30)
plt.show()

```

```python
# "Quantity Accoding The Sates"

order=df.groupby(["State"])["Quantity"].mean().sort_values(ascending=False).index
plt.figure(figsize=(6,10),dpi=400)
sns.barplot(df["Quantity"],df["State"],ci=None,order=order)

plt.title("Quantity Accoding The Sates",fontsize=15)
plt.show()


```

```python
# Sales Accoding The Sates
order=df.groupby(["State"])["Sales"].mean().sort_values(ascending=False).index
plt.figure(figsize=(10,6),dpi=400)
sns.barplot(df["State"],df["Sales"],ci=None,order=order)
plt.xticks(rotation="vertical")
plt.title("Sales Accoding The Sates",fontsize=30)
plt.show()

```

```python
# Sales Accoding The Category
order=df.groupby(["Category"])["Sales"].mean().sort_values(ascending=False).index
sns.barplot(df["Category"],df["Sales"],ci=None,order=order)
plt.title("Sales Accoding The Category",fontsize=20)
plt.show()

```

```python
# Quantity Accoding The Category
order=df.groupby(["Category"])["Quantity"].mean().sort_values(ascending=False).index
sns.barplot(df["Category"],df["Quantity"],ci=None,order=order)
plt.title("Quantity Accoding The Category",fontsize=20)
plt.show()

```

```python
sns.catplot(x="Category",y="Profit",data=df,kind='bar',ci=None,col="Region")
```

```python
sns.catplot(x="Quantity",y="Profit",data=df,kind='bar',ci=None,col="Category")
```

```python
sns.catplot(x="Quantity",y="Discount",data=df,kind='bar',ci=None,col="Category")

```

```python
plt.figure(figsize=(15,10),dpi=400)
order=df.groupby(['State'])["State"].count().sort_values().index
sns.countplot(y='State',data=df,order=order)
```
